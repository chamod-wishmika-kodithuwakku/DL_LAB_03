```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

Load the image using OpenCV

```python
import cv2
from IPython.display import Image, display

# Use the uploaded image path
image_path = 'lenna.png'
print("Uploaded image path:", image_path)

# Load and display the image
im = cv2.imread('/content/sample_data/' + image_path)

# Check if the image was loaded successfully
if im is not None:
    print("Image shape:", im.shape)
    display(Image(filename='/content/sample_data/' + image_path))
else:
    print("Failed to load the image")
```

Convert the RGB image into a grayscale image

```python
im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
im.shape
```

Define vertical and horizontal filter kernels

```python
v_kernel = np.array([
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
])
```

```python
v_kernel_new = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])
```

```python
h_kernel = np.array([
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
])
```

```python
h_kernel_new = np.array([
    [1, 1, 1],
    [0, 0, 0],
    [-1, -1, -1]
])
```

Do the filtering


```python
v_image = cv2.filter2D(im, -1, v_kernel)
h_image = cv2.filter2D(im, -1, h_kernel)
v_image_new = cv2.filter2D(im, -1, v_kernel_new)
h_image_new = cv2.filter2D(im, -1, h_kernel_new)
```

Display Results

```python
plt.imshow(im, cmap='gray')
plt.title('Original Image')

plt.figure()
plt.imshow(v_image, cmap='gray')
plt.title('Vertical Image')

plt.figure()
plt.imshow(h_image, cmap='gray')
plt.title('Horizontal Image')

plt.figure()
plt.imshow(v_image_new, cmap='gray')
plt.title('Vertical Image New')

plt.figure()
plt.imshow(h_image_new, cmap='gray')
plt.title('Horizontal Image New')
```
